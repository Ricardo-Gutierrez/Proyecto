### Módulo multiprocessing de Python

Es posible ejecutar múltiples procesos gracias los núcleos lógicos (hilos), este número puede ser conocido ejecutando el siguiente código [1]:

```python
import multiprocessing

print(f'Número de núcleos {multiprocessing.cpu_count()}')

Los procesos pueden ser agregados a un Pool de procesos para su ejecución en paralelo, una vez que un proceso termina una tarea procede a tomar otra tarea del Pool de procesos, esto se puede realizar de la siguiente manera [2]:

```python

from multiprocessing import Pool

def f(x):
    print(x*x)

if __name__ == '__main__':
    p = Pool()
    for i in range(10):
        p.apply_async(f, (i,)).get()
    p.close()
    p.join()


```

El código fuente es el siguiente, se recomienda crear un archivo .py para ejecutarlo:

```python

import time
from multiprocessing import Pool
import logging
import cv2 as cv
import os

# Relative path to the images
REL = "Multiprocessing\\"

# debugging print format
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s %(message)s',
)


def process_image(filename, threshold):  # Function to be executed in parallel
    process_id = str(os.getpid()).zfill(5)  # Get the process id
    # debugging print
    logging.debug(f'{process_id}: Processing {filename} - {threshold}')
    image = cv.imread(REL+filename)  # read image
    gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)  # convert to grayscale
    ret, thresh = cv.threshold(
        gray, threshold, 255, cv.THRESH_BINARY)  # threshold
    cv.imwrite(f'{REL}processed_{threshold}_{filename}', thresh)  # save image
    logging.debug(f'{process_id}: Finished {filename} - {threshold}')


if __name__ == '__main__':
    p = Pool()  # Create a multiprocessing Pool (if no argument is given, it will use all the cores)
    for thresh in range(0, 255, 5):  # for each threshold
        # apply the function to the image
        p.apply_async(process_image, args=(f'image.jpg', thresh))
    p.close()  # close the pool
    p.join()  # wait for all processes to finish


# Resultados

La imagen base es la siguiente:
![image info](image.jpg)

Los resultados obtenidos son los siguientes:
![image info](ss.jpg)

El resultado de la ejecución del código fuente es el siguiente:

    2022-11-09 18:45:57,347 21804: Processing image.jpg - 0
    2022-11-09 18:45:57,393 27112: Processing image.jpg - 5
    2022-11-09 18:45:57,465 11112: Processing image.jpg - 10
    2022-11-09 18:45:57,507 18060: Processing image.jpg - 15
    2022-11-09 18:45:57,513 29124: Processing image.jpg - 20
    2022-11-09 18:45:57,517 21004: Processing image.jpg - 25
    2022-11-09 18:45:57,582 25008: Processing image.jpg - 30
    2022-11-09 18:45:57,612 13980: Processing image.jpg - 35
    2022-11-09 18:45:58,014 21804: Finished image.jpg - 0
    2022-11-09 18:45:58,024 21804: Processing image.jpg - 40
    2022-11-09 18:45:58,057 27112: Finished image.jpg - 5
    2022-11-09 18:45:58,067 27112: Processing image.jpg - 45
    2022-11-09 18:45:58,127 11112: Finished image.jpg - 10
    2022-11-09 18:45:58,135 11112: Processing image.jpg - 50
    2022-11-09 18:45:58,203 18060: Finished image.jpg - 15
    2022-11-09 18:45:58,210 18060: Processing image.jpg - 55
    2022-11-09 18:45:58,214 29124: Finished image.jpg - 20
    2022-11-09 18:45:58,222 29124: Processing image.jpg - 60
    2022-11-09 18:45:58,231 21004: Finished image.jpg - 25
    2022-11-09 18:45:58,239 21004: Processing image.jpg - 65
    2022-11-09 18:45:58,301 25008: Finished image.jpg - 30
    2022-11-09 18:45:58,310 25008: Processing image.jpg - 70
    2022-11-09 18:45:58,327 13980: Finished image.jpg - 35
    2022-11-09 18:45:58,335 13980: Processing image.jpg - 75
    2022-11-09 18:45:58,831 21804: Finished image.jpg - 40
    2022-11-09 18:45:58,843 21804: Processing image.jpg - 80
    2022-11-09 18:45:58,858 27112: Finished image.jpg - 45
    2022-11-09 18:45:58,864 11112: Finished image.jpg - 50
    2022-11-09 18:45:58,867 27112: Processing image.jpg - 85
    2022-11-09 18:45:58,876 11112: Processing image.jpg - 90
    2022-11-09 18:45:59,027 13980: Finished image.jpg - 75
    2022-11-09 18:45:59,036 13980: Processing image.jpg - 95
    2022-11-09 18:45:59,098 25008: Finished image.jpg - 70
    2022-11-09 18:45:59,112 21004: Finished image.jpg - 65
    2022-11-09 18:45:59,120 21004: Processing image.jpg - 100
    2022-11-09 18:45:59,148 25008: Processing image.jpg - 105
    2022-11-09 18:45:59,213 29124: Finished image.jpg - 60
    2022-11-09 18:45:59,222 29124: Processing image.jpg - 110
    2022-11-09 18:45:59,239 18060: Finished image.jpg - 55
    2022-11-09 18:45:59,247 18060: Processing image.jpg - 115
    2022-11-09 18:45:59,551 21804: Finished image.jpg - 80
    2022-11-09 18:45:59,560 21804: Processing image.jpg - 120
    2022-11-09 18:45:59,584 11112: Finished image.jpg - 90
    2022-11-09 18:45:59,592 11112: Processing image.jpg - 125
    2022-11-09 18:45:59,758 13980: Finished image.jpg - 95
    2022-11-09 18:45:59,766 13980: Processing image.jpg - 130
    2022-11-09 18:45:59,790 27112: Finished image.jpg - 85
    2022-11-09 18:45:59,819 27112: Processing image.jpg - 135
    2022-11-09 18:45:59,850 21004: Finished image.jpg - 100
    2022-11-09 18:45:59,859 21004: Processing image.jpg - 140
    2022-11-09 18:45:59,935 29124: Finished image.jpg - 110
    2022-11-09 18:45:59,943 29124: Processing image.jpg - 145
    2022-11-09 18:45:59,967 25008: Finished image.jpg - 105
    2022-11-09 18:45:59,974 25008: Processing image.jpg - 150
    2022-11-09 18:46:00,000 18060: Finished image.jpg - 115
    2022-11-09 18:46:00,009 18060: Processing image.jpg - 155
    2022-11-09 18:46:00,348 21804: Finished image.jpg - 120
    2022-11-09 18:46:00,359 21804: Processing image.jpg - 160
    2022-11-09 18:46:00,450 11112: Finished image.jpg - 125
    2022-11-09 18:46:00,458 11112: Processing image.jpg - 165
    2022-11-09 18:46:00,529 13980: Finished image.jpg - 130
    2022-11-09 18:46:00,547 13980: Processing image.jpg - 170
    2022-11-09 18:46:00,604 21004: Finished image.jpg - 140
    2022-11-09 18:46:00,612 21004: Processing image.jpg - 175
    2022-11-09 18:46:00,640 27112: Finished image.jpg - 135
    2022-11-09 18:46:00,645 29124: Finished image.jpg - 145
    2022-11-09 18:46:00,649 27112: Processing image.jpg - 180
    2022-11-09 18:46:00,654 29124: Processing image.jpg - 185
    2022-11-09 18:46:00,811 25008: Finished image.jpg - 150
    2022-11-09 18:46:00,823 25008: Processing image.jpg - 190
    2022-11-09 18:46:00,860 18060: Finished image.jpg - 155
    2022-11-09 18:46:00,868 18060: Processing image.jpg - 195
    2022-11-09 18:46:01,093 21804: Finished image.jpg - 160
    2022-11-09 18:46:01,102 21804: Processing image.jpg - 200
    2022-11-09 18:46:01,261 11112: Finished image.jpg - 165
    2022-11-09 18:46:01,268 11112: Processing image.jpg - 205
    2022-11-09 18:46:01,332 13980: Finished image.jpg - 170
    2022-11-09 18:46:01,343 13980: Processing image.jpg - 210
    2022-11-09 18:46:01,412 29124: Finished image.jpg - 185
    2022-11-09 18:46:01,425 29124: Processing image.jpg - 215
    2022-11-09 18:46:01,453 27112: Finished image.jpg - 180
    2022-11-09 18:46:01,457 21004: Finished image.jpg - 175
    2022-11-09 18:46:01,476 27112: Processing image.jpg - 220
    2022-11-09 18:46:01,497 21004: Processing image.jpg - 225
    2022-11-09 18:46:01,653 25008: Finished image.jpg - 190
    2022-11-09 18:46:01,665 25008: Processing image.jpg - 230
    2022-11-09 18:46:01,738 18060: Finished image.jpg - 195
    2022-11-09 18:46:01,752 18060: Processing image.jpg - 235
    2022-11-09 18:46:01,919 21804: Finished image.jpg - 200
    2022-11-09 18:46:01,927 21804: Processing image.jpg - 240
    2022-11-09 18:46:01,996 11112: Finished image.jpg - 205
    2022-11-09 18:46:02,004 11112: Processing image.jpg - 245
    2022-11-09 18:46:02,143 13980: Finished image.jpg - 210
    2022-11-09 18:46:02,151 13980: Processing image.jpg - 250
    2022-11-09 18:46:02,262 29124: Finished image.jpg - 215
    2022-11-09 18:46:02,265 27112: Finished image.jpg - 220
    2022-11-09 18:46:02,373 21004: Finished image.jpg - 225
    2022-11-09 18:46:02,407 18060: Finished image.jpg - 235
    2022-11-09 18:46:02,428 25008: Finished image.jpg - 230
    2022-11-09 18:46:02,566 21804: Finished image.jpg - 240
    2022-11-09 18:46:02,745 11112: Finished image.jpg - 245
    2022-11-09 18:46:02,800 13980: Finished image.jpg - 250

## Conclusiones

El uso del módulo multiprocessing nos permite ejecutar procesos en paralelo, lo que facilita el optimizar el tiempo de ejecución de nuestro programa en comparación con ejecutarlo de forma secuencial (1 proceso a la vez). El uso de Pool permite agregar varias tareas a una especie de fila virtual que es procesada por los procesos que se encuentran en el pool, cuya cantidad depende de los hilos físicos de nuestro procesador.

## Bibliografía
1. https://docs.python.org/3/library/multiprocessing.html
2. https://superfastpython.com/multiprocessing-pool-python/
